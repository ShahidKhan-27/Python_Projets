```python
#This is my first web Scraping for 140 _90s-favorite movies.Lets do it 
from bs4 import BeautifulSoup
import requests
```

```python
url='https://editorial.rottentomatoes.com/guide/140-favorite-90s-movies/'
#GET THE HTML
html_content=requests.get(url).content
```

```python
#Parser HTML content
soup=BeautifulSoup(html_content,'lxml')
print(soup.prettify())
```

```python
#Got the title of HTML page
title=str(soup.title.text)
print(type(title))
```

```python
#text from all soup
text = soup.get_text().lstrip().replace(" ","")
print(text)
```

```python
#Get all links from page
anchortags=soup.find_all('a')
for link in anchortags:
    http=link.get('href')
    if str(http).startswith('/') is True:
        http=("https:"+link.get('href'))
    else:
        http=http
    print(http)
#statement required cause some links not startswith('https:')
```

```python
#movies release year
year=soup.find(class_='row countdown-item')
print(year)
```

```python
Movies=[]
Year=[]
Rating=[]

datas=soup.find_all('div',class_='row countdown-item')
for data in datas :
    movie_name=(data.h2.a.text)
    year=int(data.h2.span.text[1:5])
    rating=data.h2.find('span',class_='tMeterScore').text
    Movies.append(movie_name)
    Year.append(year)
    Rating.append(rating)
#made a list of all content.text to prepare a table
```

```python
Director=[]
cards=soup.find_all('div',class_="info director")
for card in cards:
    dir_name=card.a.text
    Director.append(dir_name)   
```

```python
Casts=[]
casts=soup.find_all('div',class_="info cast")
for cast in casts:
    cast_name=[]
    cast_a=cast.find_all('a')
    for cast_b in cast_a:
        cast_name.append(cast_b.text)
    Casts.append(cast_name)
```

```python
import pandas as pd
df = pd.DataFrame({'Movies_Name':Movies,'Year':Year,'Ratings':Rating,'Directors':Director,'Casting':Casts})
```
