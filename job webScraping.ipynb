```python
Job=[]
from bs4 import BeautifulSoup
import requests
import pandas as pd
html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Python&txtLocation=').text
soup = BeautifulSoup(html_text,"lxml")
jobs = soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
a=0
for job in jobs:
    a=a+1
    publish_date= job.find('span',class_='sim-posted').text.strip()
    if "few" in publish_date:
        company_name=job.find('h3',class_='joblist-comp-name').text.strip()
        skill= job.find('span',class_='srp-skills').text.strip()
        lnk=job.h2.a.get('href')
        Job.append([a,company_name,skill,lnk])
        print(Job)
df=pd.DataFrame(Job,columns=['S.no','Company Name','Skills','Links for job role'])
df.to_csv('JOBS WEB SCRAPING')
```
